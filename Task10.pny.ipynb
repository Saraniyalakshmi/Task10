```python
#.3

import random

class AudioPlayer:
    def __init__(self):
        # Initialize lists to store audios and playlists
        self.audios = []
        self.playlists = []

    def create_audio(self, url, name, genre):
        # Method to create a new audio
        audio = {"url": url, "name": name, "genre": genre, "ratings": [], "average_rating": 0}
        self.audios.append(audio)

    def create_playlist(self, name, genre):
        # Method to create a new playlist
        playlist = {"name": name, "genre": genre, "audios": [], "ratings": [], "average_rating": 0}
        self.playlists.append(playlist)

    def add_audio_to_playlist(self, audio_name, playlist_name):
        # Method to add audio to a playlist
        audio = next((audio for audio in self.audios if audio["name"] == audio_name), None)
        playlist = next((playlist for playlist in self.playlists if playlist["name"] == playlist_name), None)
        if audio and playlist:
            playlist["audios"].append(audio)

    def search_audio_by_name(self, name):
        # Method to search audio by name
        return [audio for audio in self.audios if audio["name"].lower() == name.lower()]

    def search_playlist_by_name(self, name):
        # Method to search playlist by name
        return [playlist for playlist in self.playlists if playlist["name"].lower() == name.lower()]

    def rate_audio(self, audio_name, rating):
        # Method to rate an audio
        audio = next((audio for audio in self.audios if audio["name"] == audio_name), None)
        if audio:
            audio["ratings"].append(rating)
            audio["average_rating"] = sum(audio["ratings"]) / len(audio["ratings"])

    def rate_playlist(self, playlist_name, rating):
        # Method to rate a playlist
        playlist = next((playlist for playlist in self.playlists if playlist["name"] == playlist_name), None)
        if playlist:
            playlist["ratings"].append(rating)
            playlist["average_rating"] = sum(playlist["ratings"]) / len(playlist["ratings"])

    def fetch_average_rating(self):
        # Method to fetch average rating (simulated by generating random ratings for 3 users)
        ratings = [random.randint(1, 5) for _ in range(3)]
        return sum(ratings) / len(ratings)


# Example usage:
player = AudioPlayer()

# Creating audios
player.create_audio("https://example.com/audio1.mp3", "Song 1", "Pop")
player.create_audio("https://example.com/audio2.mp3", "Song 2", "Rock")

# Creating playlists
player.create_playlist("Playlist 1", "Pop")
player.create_playlist("Playlist 2", "Rock")

# Adding audios to playlists
player.add_audio_to_playlist("Song 1", "Playlist 1")
player.add_audio_to_playlist("Song 2", "Playlist 2")

# Searching audio by name
search_result = player.search_audio_by_name("Song 1")
print("Search result:", search_result)

# Searching playlist by name
search_result = player.search_playlist_by_name("Playlist 1")
print("Search result:", search_result)

# Rating audio and playlist
player.rate_audio("Song 1", 4)
player.rate_playlist("Playlist 1", 5)

# Fetching average rating
average_rating = player.fetch_average_rating()
print("Average rating:", average_rating)
```
